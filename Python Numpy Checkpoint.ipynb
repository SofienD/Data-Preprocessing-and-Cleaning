**1- Write a Python program to convert an array to an ordinary list with the same items.**

```python
import numpy as np
A=np.array([[1,2,3],[8,5,4],[7,17,56]])
B=A.tolist()
liste_plate = []
for i in B:
    for j in i:
       liste_plate.append(j)
# Affichage
print("Initial Array:", B)
print("Ordinary List:", liste_plate)
```

**2-Write a NumPy program to compute the sum of the diagonal element of a given array.**

```python
import numpy as np
A=np.array([[1,2,3],[8,5,4],[7,17,56]])
C=A.diagonal()
somme_diagonale = 0
for i in range(0,len(C)):
  somme_diagonale = somme_diagonale + C[i]
print(somme_diagonale) 
```

**3- Given an array of your choice, get all the values higher than X :if a = [[1,2],[3,5]] and x = 2 :  then 3 and 5 are higher than 2.**

```python
import numpy as np                              # Meme code que 1 , je rajoute la partie pour la comparaison des éléments à un X donné
A=np.array([[1,2,3],[8,5,4],[7,17,56]])
B=A.tolist()
liste_plate=[]
for i in B:
    for j in i:
       liste_plate.append(j)                                

list=[]                                     # Définition d'une liste vide qui contiendra les éléments supérieurs à x
x=int(input())
for c in liste_plate:
   if c>x:
      list.append(c)

print(list)
```
