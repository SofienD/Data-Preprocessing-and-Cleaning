**3. Write a Python program to read last n lines of a file**

```python
f.close()
def derniers_n_lignes(fname, N): 
  with open(fname) as file:
    for line in (file.readlines() [-N:]): 
            print(line, end ='')
```

```python
derniers_n_lignes("python.txt",3)
```

**4. Write a Python program that takes a text file as input and returns the number of words of a given text file.**

```python
def calculateur_mots(fname): 
    f=open(fname,'r',encoding='utf-8')
    text=f.read()
    list_of_words = text.split()
    l=len(list_of_words)
    return("le nombre de mots dans le texte est:", l)
```

```python
calculateur_mots("python.txt")
```
